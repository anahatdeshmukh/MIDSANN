```python
# Assignment 5 - K-Means Clustering

import numpy as np

# Given points
points = {
    "P1": [0.1, 0.6],
    "P2": [0.15, 0.71],
    "P3": [0.08, 0.9],
    "P4": [0.16, 0.85],
    "P5": [0.2, 0.3],
    "P6": [0.25, 0.5],
    "P7": [0.24, 0.1],
    "P8": [0.3, 0.2]
}

# Initial centroids
m1 = np.array(points["P1"])   # Cluster C1
m2 = np.array(points["P8"])   # Cluster C2

C1 = []
C2 = []

# Euclidean Distance Function
def distance(a, b):
    return np.sqrt(np.sum((a - b) ** 2))

# Assign points to clusters
for name, point in points.items():

    point = np.array(point)

    d1 = distance(point, m1)
    d2 = distance(point, m2)

    if d1 < d2:
        C1.append(point)
        print(name, "belongs to Cluster C1")
    else:
        C2.append(point)
        print(name, "belongs to Cluster C2")

# Updated centroids
new_m1 = np.mean(C1, axis=0)
new_m2 = np.mean(C2, axis=0)

print("\nUpdated Centroid m1:", new_m1)
print("Updated Centroid m2:", new_m2)

print("\nPopulation around m2:", len(C2))
```
