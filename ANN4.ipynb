```python
# Assignment 4 - Demonstration of ART (Adaptive Resonance Theory)

import numpy as np

# Input patterns
patterns = np.array([
    [1, 1, 0, 0],
    [1, 0, 0, 0],
    [0, 0, 1, 1],
    [0, 0, 1, 0]
])

# Simple ART clustering
clusters = []

threshold = 0.5

for p in patterns:

    assigned = False

    for i, c in enumerate(clusters):

        similarity = np.sum(p == c) / len(p)

        if similarity >= threshold:
            print("Pattern", p, "belongs to Cluster", i + 1)
            assigned = True
            break

    if not assigned:
        clusters.append(p)
        print("Pattern", p, "creates new Cluster", len(clusters))

# Display clusters
print("\nFinal Clusters:")
for i, c in enumerate(clusters):
    print("Cluster", i + 1, ":", c)
```
