```python
import numpy as np
import matplotlib.pyplot as plt

# Training data
X = np.array([[1,2],[2,3],[3,6],[4,5]])
y = np.array([0,0,1,1])

# Perceptron training
w = np.zeros(2)
b = 0
lr = 1

for _ in range(10):
    for i in range(len(X)):

        pred = 1 if np.dot(X[i],w)+b >= 0 else 0

        error = y[i] - pred

        w = w + lr * error * X[i]
        b = b + lr * error

# Plot points
for i in range(len(X)):

    if y[i] == 0:
        plt.scatter(X[i][0], X[i][1],
                    color='black', marker='s', s=100)

    else:
        plt.scatter(X[i][0], X[i][1],
                    color='black', marker='o', s=100)

# Decision boundary
x = np.linspace(0,7,100)
y_line = -(w[0]*x + b)/w[1]

plt.plot(x, y_line, 'k--', label="Decision Boundary")

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Perceptron Decision Boundary")
plt.legend()
plt.grid()

plt.show()

# Test new data
test_points = np.array([[3,2],[4,5]])

print("\nTest Predictions:\n")

for point in test_points:

    pred = 1 if np.dot(point,w)+b >= 0 else 0

    print(point,
          "classified as:",
          "Class 1" if pred==1 else "Class 0")
```
