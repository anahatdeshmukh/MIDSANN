```python
import numpy as np

# Sigmoid activation
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of sigmoid
def sigmoid_derivative(x):
    return x * (1 - x)

# XOR input
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

# XOR output
y = np.array([
    [0],
    [1],
    [1],
    [0]
])

# Random initialization
np.random.seed(1)

w1 = np.random.rand(2,2)
w2 = np.random.rand(2,1)

b1 = np.zeros((1,2))
b2 = np.zeros((1,1))

lr = 0.5

# ---- Training ----
for epoch in range(10000):

    # Forward pass
    hidden = sigmoid(np.dot(X, w1) + b1)
    output = sigmoid(np.dot(hidden, w2) + b2)

    # Error
    error = y - output

    # PRINT ERROR EVERY 1000 EPOCHS
    if epoch % 1000 == 0:
        print("Error:", np.mean(np.abs(error)))

    # Backpropagation (3-line version)
    d_output = error * sigmoid_derivative(output)

    error_hidden = d_output.dot(w2.T)
    d_hidden = error_hidden * sigmoid_derivative(hidden)

    # Update weights
    w2 += hidden.T.dot(d_output) * lr
    w1 += X.T.dot(d_hidden) * lr

    b2 += np.sum(d_output, axis=0, keepdims=True) * lr
    b1 += np.sum(d_hidden, axis=0, keepdims=True) * lr


# ---- Output ----
print("Final Binary Output:")

hidden = sigmoid(np.dot(X, w1) + b1)
output = sigmoid(np.dot(hidden, w2) + b2)

binary = (output >= 0.5).astype(int)

for i in range(len(X)):
    print(X[i], "->", binary[i][0])
```
