```python
# Assignment 5 - Perceptron Learning Law with Decision Region

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron

# Input data
X = np.array([
    [1, 1],
    [2, 1],
    [2, 3],
    [3, 2],
    [6, 5],
    [7, 7],
    [8, 6],
    [9, 7]
])

# Output labels
y = np.array([0, 0, 0, 0, 1, 1, 1, 1])

# Train perceptron model
model = Perceptron()
model.fit(X, y)

# Plot data points
for i in range(len(X)):
    if y[i] == 0:
        plt.scatter(X[i][0], X[i][1], marker='o')
    else:
        plt.scatter(X[i][0], X[i][1], marker='x')

# Decision boundary
w = model.coef_[0]
b = model.intercept_[0]

x_points = np.linspace(0, 10, 100)
y_points = -(w[0] * x_points + b) / w[1]

plt.plot(x_points, y_points)

# Labels
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Perceptron Decision Region")

# Show graph
plt.show()
```
