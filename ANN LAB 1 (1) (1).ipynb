```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-10,10,100)

def sigmoid(x):
    return 1/(1+np.exp(-x))

def tanh(x):
    return np.tanh(x)

def Relu(x):
    return np.maximum(0,x)

def leaky_relu(x):
    return np.where(x>0,x,0.1*x)

#sigmoid
plt.figure(figsize=(5,4))
plt.plot(x,sigmoid(x))
plt.title("Sigmoid Graph")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()

#Tanh
plt.figure(figsize=(5,4))
plt.plot(x,tanh(x))
plt.title("tanh graph")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()

#Relu
plt.figure(figsize=(5,4))
plt.plot(x,Relu(x))
plt.title("Relu graph")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()

#leky-relu
plt.figure(figsize=(5,4))
plt.plot(x,leaky_relu(x))
plt.title("leaky-relu")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```
